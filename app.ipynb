```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15


# Example usage
t = Temperature(25)
print(f"Celsius: {t.celsius}")
print(f"Fahrenheit: {t.to_fahrenheit()}")
print(f"Kelvin: {t.to_kelvin()}")
```

```python
class InterestCalculator:
    def __init__(self, principal, rate, time):
        self.principal = principal
        self.rate = rate      # in percentage per annum
        self.time = time      # in years

    def simple_interest(self):
        return (self.principal * self.rate * self.time) / 100

    def compound_interest(self, n=1):
        """n = number of times interest is compounded per year"""
        amount = self.principal * (1 + (self.rate / (100 * n))) ** (n * self.time)
        return amount - self.principal


# Example usage
calc = InterestCalculator(principal=10000, rate=5, time=2)
print(f"Simple Interest: {calc.simple_interest():.2f}")
print(f"Compound Interest (annually): {calc.compound_interest():.2f}")
print(f"Compound Interest (quarterly): {calc.compound_interest(n=4):.2f}")
```

```python
class CoffeeMachine:
    def __init__(self, water, coffee, milk):
        self.water = water
        self.coffee = coffee
        self.milk = milk

    def make_latte(self):
        required_water = 200
        required_coffee = 20
        required_milk = 150

        if (self.water >= required_water and
                self.coffee >= required_coffee and
                self.milk >= required_milk):
            self.water -= required_water
            self.coffee -= required_coffee
            self.milk -= required_milk
            print("Latte made successfully! ☕")
            print(f"Remaining -> Water: {self.water}ml, Coffee: {self.coffee}g, Milk: {self.milk}ml")
        else:
            print("Not enough resources to make a latte!")
            missing = []
            if self.water < required_water:
                missing.append(f"Water (need {required_water - self.water}ml more)")
            if self.coffee < required_coffee:
                missing.append(f"Coffee (need {required_coffee - self.coffee}g more)")
            if self.milk < required_milk:
                missing.append(f"Milk (need {required_milk - self.milk}ml more)")
            print("Missing: " + ", ".join(missing))


# Example usage
machine = CoffeeMachine(water=300, coffee=100, milk=200)
machine.make_latte()   # Should succeed
machine.make_latte()   # Should fail (insufficient milk left)
```

```python
class Vehicle:
    def __init__(self, name, max_speed):
        self.name = name
        self.max_speed = max_speed

    def display(self):
        print(f"Vehicle Name: {self.name}")
        print(f"Max Speed: {self.max_speed} km/h")


class Bus(Vehicle):
    def __init__(self, name, max_speed, seating_capacity):
        super().__init__(name, max_speed)
        self.seating_capacity = seating_capacity
        self.reserved_seats = 0

    def display(self):
        super().display()
        print(f"Seating Capacity: {self.seating_capacity}")
        print(f"Reserved Seats Today: {self.reserved_seats}")

    def reserve_seats(self, num_seats):
        available = self.seating_capacity - self.reserved_seats
        if num_seats <= 0:
            print("Number of seats must be positive.")
        elif num_seats <= available:
            self.reserved_seats += num_seats
            print(f"{num_seats} seat(s) reserved successfully!")
            print(f"Available seats remaining today: {self.seating_capacity - self.reserved_seats}")
        else:
            print(f"Cannot reserve {num_seats} seat(s). Only {available} seat(s) available today.")


# Example usage
bus = Bus(name="Volvo Express", max_speed=100, seating_capacity=40)
bus.display()
print()
bus.reserve_seats(15)
bus.reserve_seats(30)  # Should fail — not enough seats left
print()
bus.display()
```
