```python
Name= input("enter your name:")
Clas=int(input("enter your class:"))
Subject1=float(input("enter your marks subject1:"))
Subject2=float(input("enter your marks subject2:"))
Subject3=float(input("enter your marks subject3:"))
Sum= Subject1+ Subject2+ Subject3
Percantage= Sum/300*100 

if Percantage>=80:
    grade = "A"
elif Percantage>=70:
    grade = "B"
elif Percantage>=60:
    grade = "c" 
elif Percantage>=50:
    grade = "D" 
else :
    grade ="fail"       

print(f"\n Name: {Name}\n Class:{Clas}\n Subject1: {Subject1}\n Subject2: {Subject2}\n Subject : {Subject3} \n Sum: {Sum} \n Percantage: {Percantage} \n grade: {grade}")

```

```python
n = int(input("enter the number:"))

temp = n
rev = 0

while n>0:
    rem = n % 10
    rev = rev * 10 + rem 
    n = n // 10

print("Enter number =" , temp)
print("reverse number =", rev)

if temp == rev:
        print("palindrome number")
else:
        print("not A palindrome number")
```
