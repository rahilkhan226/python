```python
def maximum(a, b, c):
    if a>=b and a>=c:
        return a

    elif b>=a and b>=c:
        return b

    else:
        return c



print("maximum:",maximum(10,20,30))
    

```

```python
def function(lst):
    my_newlist= []

    for item in lst:
        if item not in my_newlist:
            my_newlist.append(item)

    return my_newlist
        
print(function([1,1,2,3,4,5,6,6,7,7]))
    
```

```python
def function(lst):
    result = 2

    for num in lst:
        result=result*num

    return result
print(function([40,40]))
```

```python
def function(n):
    fact =1
    for i  in range (1, n+1):
        fact = fact*i
    
    return fact
num=int(input("enter an value:"))
print("factorial:",function(num))
```

```python
def prime(n):
    for i in range(2,n):
        if n % i == 0:
            print("not prime")
            return
        
    print("prime")

prime(8)
    
```

```python
def function(s):
    upper = 0
    lower = 0

    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    print("Uppercase Letters:", upper)
    print("Lowercase Letters:", lower)

function("Hello World")
```
