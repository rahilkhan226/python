```python
t = (10,20,30,40,40)
total=0
highest=t[0]
lowest=t[0]

for num in t:
    total = total+num
    
    if num > highest:
       highest=num
    
    if num<lowest:
       lowest=num

print("Total =", total)
print("Highest Value =", highest)
print("Lowest Value =", lowest)
```
