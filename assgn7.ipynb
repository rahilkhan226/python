```python
import csv

while True:
    print("1. Add Contact")
    print("2. View Contact")
    print("3. Exit")

    ch = input("Enter Choice: ")

    if ch == "1":
        name = input("Enter Name: ")
        mobile = input("Enter Mobile: ")
        email = input("Enter Email: ")

        data = [name, mobile, email]

        with open("address_book.csv", "a", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(data)

        print("Contact Added")

    elif ch == "2":
        with open("address_book.csv", "r") as f:
            reader = csv.reader(f)

            for row in reader:
                print(row)

    elif ch == "3":
        break

    else:
        print("Invalid Choice")
```
