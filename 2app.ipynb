```python
# Assignment 2 - Question 1

name = input("Enter student name: ")
student_class = input("Enter class: ")

print("Enter marks for 5 subjects:")
marks = []
for i in range(1, 6):
    mark = float(input(f"  Subject {i}: "))
    marks.append(mark)

total = sum(marks)
percentage = (total / 500) * 100

print("\n--- Result ---")
print(f"Name       : {name}")
print(f"Class      : {student_class}")
print(f"Total Marks: {total}/500")
print(f"Percentage : {percentage:.2f}%")
```

```python
# Assignment 2 - Question 2

s1 = input("Enter first string: ")
s2 = input("Enter second string: ")

s = s1 + s2
print(f"\nConcatenated string: '{s}'")
print("-" * 40)

print(f"lower()      : {s.lower()}")
print(f"upper()      : {s.upper()}")
print(f"title()      : {s.title()}")
print(f"swapcase()   : {s.swapcase()}")
print(f"capitalize() : {s.capitalize()}")
print(f"casefold()   : {s.casefold()}")
print(f"center(40)   : '{s.center(40)}'")
print(f"count('a')   : {s.count('a')}")
print(f"endswith('n'): {s.endswith('n')}")
print(f"find('e')    : {s.find('e')}")
print(f"isalnum()    : {s.isalnum()}")
print(f"isdigit()    : {s.isdigit()}")
print(f"isnumeric()  : {s.isnumeric()}")
print(f"isspace()    : {s.isspace()}")
print(f"replace('a','@'): {s.replace('a', '@')}")
```

```python
# Assignment 2 - Question 3

x = 10
print(f"Initial value: x = {x}")

x += 5;  print(f"x += 5  → x = {x}")   # Addition
x -= 3;  print(f"x -= 3  → x = {x}")   # Subtraction
x *= 2;  print(f"x *= 2  → x = {x}")   # Multiplication
x /= 4;  print(f"x /= 4  → x = {x}")   # Division
x //= 2; print(f"x //= 2 → x = {x}")   # Floor division
x **= 3; print(f"x **= 3 → x = {x}")   # Exponentiation
x %= 5;  print(f"x %%= 5  → x = {x}")  # Modulus

y = 0b1111  # 15 in binary
print(f"\nBitwise (y = {y}):")
y &= 0b1010; print(f"y &= 0b1010 → y = {y}")   # AND
y |= 0b0101; print(f"y |= 0b0101 → y = {y}")   # OR
y ^= 0b1111; print(f"y ^= 0b1111 → y = {y}")   # XOR
y >>= 1;     print(f"y >>= 1     → y = {y}")   # Right shift
y <<= 2;     print(f"y <<= 2     → y = {y}")   # Left shift
```
